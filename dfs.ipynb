```python
import networkx as nx
import matplotlib.pyplot as plt


def dfs(visited, graph, node, path):
    if node not in visited:
        print(node, end=" ")
        visited.add(node)
        path.append(node)  

        for neighbour in graph[node]:
            dfs(visited, graph, neighbour, path)

def show_graph(graph):
    G = nx.Graph()
    
    for node1 in graph:
        G.add_node(node1)
        for node2 in graph[node1]:
            G.add_edge(node1, node2)

    nx.draw(G, with_labels=True, node_color='red', node_size=1500, font_size=20, font_color='black')
    plt.show()

visited = set()  
graph = dict()
path = []  

n = int(input("Enter number of nodes: ")) 
for i in range(1, n + 1):
    edges = int(input("Enter number of edges for node {}: ".format(i)))  
    graph[i] = []  

    for j in range(1, edges + 1):  
        node = int(input("Enter edge {} for node {}: ".format(j, i)))
        graph[i].append(node)

print("The following is DFS: ")
dfs(visited, graph, 1, path)
print("\nPath: ", path)  
show_graph(graph)
```
