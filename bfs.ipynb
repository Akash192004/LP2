```python
import networkx as nx
import matplotlib.pyplot as plt


def bfs(visited, graph, node, queue, path):
    visited.add(node)
    queue.append(node)
    path[node] = [node]  

    while queue:
        s = queue.pop(0)
        print(s, end=" ")

        for neighbour in graph[s]:
            if neighbour not in visited:
                visited.add(neighbour)
                queue.append(neighbour)
               
                path[neighbour] = path[s] + [neighbour]

def show_graph(graph):
    G = nx.Graph()
   
    for node1 in graph:
        G.add_node(node1)
        for node2 in graph[node1]:
            G.add_edge(node1, node2)

    nx.draw(G, with_labels=True, node_color='red', node_size=1500, font_size=20, font_color='black')
    plt.show()

visited = set()  
queue = [] 
graph = dict()
path = dict()

n = int(input("Enter number of nodes:"))  
for i in range(1, n + 1):
    edges = int(input("Enter number of edges for node {}: ".format(i)))  
    graph[i] = [] 
    path[i] = []  

    for j in range(1, edges + 1): 
        node = int(input("Enter edge {} for node {}: ".format(j, i)))
       
        graph[i].append(node)

print("The following is BFS: ")
bfs(visited, graph, 1, queue, path)
print("\nPaths from start node to each node:")
for node, p in path.items():
    print("Node {}: {}".format(node, p))

show_graph(graph)
```
